```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
hashtags = ["SUGA", "JIMIN", "JIN", "JHOPE", "JUNGKOOK", "TAEHYUNG", "BTSJIN", "kpop", "ATEEZ", "jimin"]
count = [45, 27, 6, 6, 5, 5, 5, 4, 4, 3]

fig, ax = plt.subplots()
ax.pie(count, labels=hashtags, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Hashtags after #BTS')

plt.show()
```
